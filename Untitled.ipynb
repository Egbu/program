```python
#works with the 4-hour scheduler. gets data every 4 hours
def four_hour():
    currency_pairs = ['GBPUSD','EURUSD','AUDUSD','GBPNZD','USDCAD']
    dates = now
    array = []
    url = "https://marketdata.tradermade.com/api/v1/minute_historical"
    for i in fx:
        for date in dates: 
            querystring = {"currency":i,"date_time":date, "api_key":api_key}
            response = requests.get(url, params=querystring)
            array.append(response.json())

    df = pd.DataFrame(array).drop(['endpoint'],axis=1)
    df = df[['currency','close','high','low','open','date_time','request_time']]
    return df
```

```python
#works with the 24-hour scheduler. Gets data every 24 hours

def twenty_four():
    fx = ['GBPUSD','EURUSD','AUDUSD','GBPNZD','USDCAD']
    dates = now - t_delta
    array = []
    url = "https://marketdata.tradermade.com/api/v1/historical"
    for i in fx:
        for date in dates: 
            querystring = {"currency":i,"date_time":date, "api_key":api_key}
            response = requests.get(url, params=querystring)
            array.append(response.json())

    df = pd.DataFrame(array).drop(['endpoint'],axis=1)
    df = df[['currency','close','high','low','open','date_time','request_time']]
    return df
```
